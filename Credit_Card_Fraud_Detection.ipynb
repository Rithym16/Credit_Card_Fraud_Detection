```python
#now we will do some more processing(exploration) just to get more insights of the data
#mainly using the class variable value
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

# From the above description we come to know that fraud transaction has more mean than legit transaction but the maximum amount is very low when compared with legit transaction.

```python
credit_card_data.groupby('Class').mean()
```

the above helps in comparing the mean values of both the class

Dealing with unbalanced data

we will use undersampling first

```python
legit_sample=legit.sample(n=492) #this function will randomly pick 492 values from legit transaction
```

making a new data set we will concatenate the two dataframes


```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) # axis is used to add values in a specific way
#if axis is zero then legit_sample data is added first then fraud data is added
```

```python
new_dataset.tail()
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

after using the mean function we can see that the mean values of both the classes are still very different therefore the characteristics of dataset is not changed which is important.

```python
new_dataset.groupby('Class').mean()
```

the above sample is good.

Feeding the data to machine learning model

Splitting the data into targets and features
Target-0 or 1

```python
X=new_dataset.drop('Class',axis=1) #X represents features
```

```python
print(X)
```

```python
Y=new_dataset['Class']
```

```python
print(Y)
```

Splitting the data into Training Data and Testing data

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING

LOGISTIC REGRESSION

```python
model=LogisticRegression()
```

```python
#training the model with training data
```

```python
model.fit(X_train, Y_train)
```

MODEL EVALUATION


Acurracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Training data accuracy', training_data_accuracy)
```

The accuracy of our model for training data comes out to be 93.26%

```python
#ACCURACY SCORE ON TESTING DATA IS MORE IMPORTANT
```

Accuracy Score on Testing Data

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data', testing_data_accuracy)
```

For testing data the accuracy score is 90% which is resonable
So out of 100 inputs our model is able to predictict 90 correct outputs

IF THE ACCURACY SCORE IS NOT SIMILAR FOR TEST AND TRAIN DATA THEN THE MODEL IS OVERFITTED OR UNDERFITTED
